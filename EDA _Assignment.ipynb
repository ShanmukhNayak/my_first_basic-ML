# Exploratory Data Analysis on Dimonds

Importing all the required libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

### Reading a csv file - 'diamonds.csv'

```python
Diamonds=pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv')
```

### Line Chart

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=Diamonds['color'],y=Diamonds['price'])
plt.xlable='color'
plt.ylable='price'
```

### Bar Chart 

It shows the relation between price and cut of the diamond

```python
plt.figure(figsize=(10,5))
plt.title("Relation between carats and price of Diamonds")
sns.barplot(x=Diamonds['cut'],y=Diamonds['price'])
xlabel='cut'
ylabel='price'
```

### Box plot

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='color',y='carat',data=Diamonds,palette='winter')
xlabel='color'
ylabel='carat'
```

### Scatter Plots

Relaion between Depth and Table of diamonds.

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='depth',y='table',hue='cut',data=Diamonds)
xlabel='depth'
ylabel='table'
```

The stregth of the relationship of the above plot can be described below.

### Regression lines for scatter plot

```python
plt.figure(figsize=(10,5))
sns.lmplot(x='depth',y='table',hue='cut',data=Diamonds)
xlabel='depth'
ylabel='table'
```

### Histograms

```python
plt.figure(figsize=(10,5))
sns.histplot(Diamonds['carat'],kde=False,color='red')
plt.ylable='Count'
plt.xlable='carat'
```

```python
Diamonds['price'].hist()
```

### Kernal Density Estimation of 'table' column

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data=Diamonds['table'],shade='True')
xlabel='table'
ylabel='Density'
```

### 2D KDE plot (Joint Plot)

```python
plt.figure(figsize=(10,5))
sns.jointplot(x=Diamonds['table'],y=Diamonds['price'],kind='kde')
plt.xlable='table'
plt.ylable='price'
```

In this way we can use so many plots to explore and analyse data
